```python
def get_grade(score):
    if score >= 90:
        return 'A'
    elif score >= 85:
        return 'B'
    elif score >= 80:
        return 'C'
    elif score >= 70:
        return 'D'
    else:
        return 'F'

scores = {}

file_path = "/scores.txt"
with open(file_path, 'r') as f:
    for line in f:
        parts = line.strip().split()
        if len(parts) == 2:
            name, score = parts
            scores[name] = int(score)

print("Name\tScore\tGrade")

for name, score in sorted(scores.items(), key=lambda item: item[1]):
    grade = get_grade(score)
    print(f"{name}\t{score}\t{grade}")

avg_score = sum(scores.values()) / len(scores)
print(f"\nAverage Score: {avg_score:.2f}")
```

```python
with open("/alt1.txt", "r") as infile:
    words = infile.read().split()  # read all words (split by whitespace)

a_words = []
b_words = []

# Separate words starting with 'a' or 'b'
for word in words:
    if word.lower().startswith('a'):
        a_words.append(word)
    elif word.lower().startswith('b'):
        b_words.append(word)

# Sort the lists
a_words.sort(reverse=True)  # reverse order (Z → A)
b_words.sort(reverse=True)  # reverse order (Z → A)

# Write 'a' words
with open("awords.txt", "w") as a_file:
    for word in a_words:
        a_file.write(word + "\n")

# Write 'b' words
with open("bwords.txt", "w") as b_file:
    for word in b_words:
        b_file.write(word + "\n")

print("Files created: awords.txt and bwords.txt")
```

```python
with open("/fruits.txt", "r") as f:
    fruits = [line.strip() for line in f if line.strip()]  # remove empty lines

# Case-sensitive sort (default)
print("1. Sorted by name (case-sensitive):")
for fruit in sorted(fruits):
    print(fruit)

# Case-insensitive sort
print("\n2. Sorted by name (case-insensitive):")
for fruit in sorted(fruits, key=str.lower):
    print(fruit)

#Sorted by length, then by name
print("\n3. Sorted by length of name, then by name:")
for fruit in sorted(fruits, key=lambda x: (len(x), x)):
    print(fruit)

#Sorted by 2nd letter, then by 1st letter
print("\n4. Sorted by 2nd letter, then by 1st letter:")
for fruit in sorted(fruits, key=lambda x: (x[1] if len(x) > 1 else '', x[0])):
    print(fruit)
```

```python
# sort_presidents.py

# Read file and split lines
presidents = []
with open("/presidentslist.txt", "r") as f:
    for line in f:
        parts = line.strip().split()
        if len(parts) == 3:
            presidents.append(parts)
        else:
            print(f"Skipping line with unexpected format: {line.strip()}")


# Sort by last name, then first name
sorted_presidents = sorted(presidents, key=lambda x: (x[1], x[0]))

# Print header
print(f"{'First Name':<15}{'Last Name':<20}{'State of Birth':<20}")
print("-" * 55)

# Print each row with alignment
for first, last, state in sorted_presidents:
    print(f"{first:<20}{last:<25}{state:<20}")
```

```python
import re

filename = input("Enter filename: ")

word_counts = {}

split_pattern = r"[^\w']+"

try:
    with open(filename, 'r') as f:
        for line in f:
            words = re.split(split_pattern, line.strip().lower())
            for word in words:
                if word:
                    word_counts[word] = word_counts.get(word, 0) + 1

    print("\nWord Frequency List:\n")
    for word in sorted(word_counts):
        print(f"{word:<15} {word_counts[word]}")

except FileNotFoundError:
    print("Error: File not found. Please check the filename and path.")
```

```python
import os

# ----- Define the Knight class -----
class Knight:
    def __init__(self, name, data_file="/knights (1).txt"):
        self._name = name
        self._title = None
        self._favorite_color = None
        self._quest = None
        self._comment = None

        # Check if the data file exists
        if not os.path.exists(data_file):
            raise FileNotFoundError(f"{data_file} not found.")

        # Read the knight data from file
        with open(data_file, "r") as f:
            for line in f:
                parts = [p.strip() for p in line.strip().split(",")]
                if len(parts) != 5:
                    continue  # skip invalid lines
                n, t, color, quest, comment = parts
                if n.lower() == name.lower():
                    self._title = t
                    self._favorite_color = color
                    self._quest = quest
                    self._comment = comment
                    break

        # If knight not found
        if self._title is None:
            raise ValueError(f"No knight found with name '{name}'")

    # Read-only properties
    @property
    def name(self):
        return self._name

    @property
    def title(self):
        return self._title

    @property
    def favorite_color(self):
        return self._favorite_color

    @property
    def quest(self):
        return self._quest

    @property
    def comment(self):
        return self._comment

if __name__ == "__main__":
    try:
        k = Knight("Arthur")
        print("Name:", k.name)
        print("Title:", k.title)
        print("Favorite Color:", k.favorite_color)
        print("Quest:", k.quest)
        print("Comment:", k.comment)
    except Exception as e:
        print(e)
```

```python
import os

# ----- Define the Knight class -----
class Knight:
    def __init__(self, name, data_file="/knights (1).txt"):
        self._name = name
        self._title = None
        self._favorite_color = None
        self._quest = None
        self._comment = None

        # Check if the data file exists
        if not os.path.exists(data_file):
            raise FileNotFoundError(f"{data_file} not found.")

        # Read the knight data from file
        with open(data_file, "r") as f:
            for line in f:
                parts = [p.strip() for p in line.strip().split(",")]
                if len(parts) != 5:
                    continue  # skip invalid lines
                n, t, color, quest, comment = parts
                if n.lower() == name.lower():
                    self._title = t
                    self._favorite_color = color
                    self._quest = quest
                    self._comment = comment
                    break

        # If knight not found
        if self._title is None:
            raise ValueError(f"No knight found with name '{name}'")

    # Read-only properties
    @property
    def name(self):
        return self._name

    @property
    def title(self):
        return self._title

    @property
    def favorite_color(self):
        return self._favorite_color

    @property
    def quest(self):
        return self._quest

    @property
    def comment(self):
        return self._comment


# ----- Application code -----
if __name__ == "__main__":
    print("Enter knight names separated by commas (e.g., Arthur,Lancelot,Gawain):")
    input_names = input().split(",")

    for name in input_names:
        name = name.strip()  # remove extra spaces
        try:
            k = Knight(name)
            print(f"\n{ k.title } { k.name }")
            print(f"Name: {k.title} {k.name}")
            print(f"Favorite Color: {k.favorite_color}")
            print(f"Quest: {k.quest}")
            print(f"Comment: {k.comment}")
        except Exception as e:
            print(f"Error for '{name}': {e}")
```
